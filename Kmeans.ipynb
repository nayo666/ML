# Kmeans aplicaci√≥n

```python
import numpy as np
from skimage.io import imread
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
img = imread("01.jpg")
plt.figure(figsize=(10, 10))
plt.imshow(img)
plt.show()
```

```python
s = img.shape
X = img.reshape(s[0] * s[1], s[2])
print(X.shape)
```

# Separar el objeto del fondo

```python
K = 2
km = KMeans(K)
X_km = km.fit_predict(X)
```

```python
Y = X_km.reshape(s[0],s[1])
plt.figure(figsize=(10, 10))
plt.imshow((Y.astype("uint8")))
plt.show()
```

# Aplicar K-means a los puntos de objetos

```python
coin_points = np.where(Y>0)
coin_points = np.concatenate([coin_points], axis=1).transpose()
coin_points.shape
```

```python
coin_points
```

```python
K = 3
km2 = KMeans(K)
km2.fit(coin_points)
coins = km2.cluster_centers_
coins
```

```python
plt.imshow(img)
plt.scatter(coins[:, 1], coins[:, 0], c="r", marker='x', s=100)
plt.show()
```

# M√°s ejemplos

```python
img2 = imread("02.jpg")
s = img2.shape
X = img2.reshape(s[0] * s[1], s[2])
K = 3
km = KMeans(K)
pred = km.fit_predict(X).reshape(s[0], s[1])
plt.imshow(pred)
plt.show()
```

```python
monedas = 3
coin_points = np.where(pred>0)
coin_points = np.concatenate([coin_points], axis=1).transpose()
km2 = KMeans(monedas)
km2.fit(coin_points)
coins = km2.cluster_centers_
plt.imshow(img2)
plt.scatter(coins[:, 1], coins[:, 0], c="r", marker='x', s=100)
plt.show()
```
