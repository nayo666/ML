# Regularización

```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.linear_model import Ridge
from sklearn.preprocessing import PolynomialFeatures
```

```python
X = np.arange(0, 10, .5).reshape(20, 1)
y = X * np.sin(X)
plt.scatter(X, y)
plt.show()
```

```python
orders = [2, 3, 4, 5, 6,7,8]
colors = ["r", "g", "y", "b", "m","black","orange"]
plt.figure(figsize=(20,10))
plt.scatter(X, y)
for i, o in enumerate(orders):
    # Código aquí
    poly = PolynomialFeatures(o)
    X_ext=poly.fit_transform(X)
    clf = Ridge(alpha=0.0, normalize=True)
    clf.fit(X_ext, y)
    pred = np.matmul(clf.coef_,X_ext.T ).T
    plt.plot(X, pred, c=colors[i], label="Grado: %d" %o)
plt.legend(loc='lower left')
plt.show()
```
