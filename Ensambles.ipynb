## Obtener el mejor macro F1 posible

Los datos ya están listos, sólo hay que agregar los clasificadores y probar (se pueden usar PolynomialFeatures)

```python
from sklearn.svm import SVC
import pandas as pd
import numpy as np
from sklearn import datasets
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import StackingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
```

# Los datos

Solo debe poner la dirección del archivo heart.csv

```python
dataset = pd.read_csv("/home/noxd/Doc/Machine Learning/datos/heart.csv", header=0)
complete_set = pd.get_dummies(dataset, columns=["cp","restecg", "slope", "thal"], prefix=["cp", "restecg", "slope", "thal"])
complete_set.head()
```

```python
X = complete_set.drop("target", axis=1).to_numpy()
y = complete_set["target"].to_numpy()
print(X.shape, y.shape)
```

```python
scaler = StandardScaler()
scaler.fit(X)
X_s = scaler.transform(X)
X_train, X_test, y_train, y_test = train_test_split(X_s, y, test_size=0.3, random_state=42)
```

# Instrucciones:
Llenar los estimators con clasificadores como SVM, Logistic Regression y KNN para obtener la mejor predicción posible en el Test.

Elegir un estimador final.

```python
for c in range(1,3):
    for i in range(1,3):
        for n in range(1,3):
            for d in range(1,4):
                pipe1 = Pipeline([('scaler', StandardScaler()), ('svc linear', SVC(C=c, kernel='linear'))])
                pipe2 = Pipeline([('scaler', StandardScaler()), ('knn', KNeighborsClassifier(n_neighbors=n))])
                pipe3 = Pipeline([('scaler', StandardScaler()), ('svc rbf', SVC(C=i, kernel='rbf'))])

                final_estimator = Pipeline([('poly_f', PolynomialFeatures(degree=d)),  ('LR', LogisticRegression(max_iter=1000))])

                estimators = [('SVM linear', pipe1),
                              ('KNN', pipe2),
                              ('SVM RBF', pipe3)]
                stacking = StackingClassifier(estimators=estimators, final_estimator=final_estimator)
                stacking.fit(X_train, y_train)
                pred = stacking.predict(X_test)
                print("Resultados para SVC lineal con c de",c,", svc rdf de", i, ", KN con",n, "vecinos y grado polinomial", d)
                print(classification_report(y_test, pred, digits=4))
```

# Ver los resultados (para SVC lineal con c de 2 , svc rdf de 2 , KN con 1 vecinos y grado polinomial 1)

```python
pipe1 = Pipeline([('scaler', StandardScaler()), ('svc linear', SVC(C=2, kernel='linear'))])
pipe2 = Pipeline([('scaler', StandardScaler()), ('knn', KNeighborsClassifier(n_neighbors=1))])
pipe3 = Pipeline([('scaler', StandardScaler()), ('svc rbf', SVC(C=2, kernel='rbf'))])

final_estimator = Pipeline([('poly_f', PolynomialFeatures(degree=1)),  ('LR', LogisticRegression(max_iter=1000))])

estimators = [('SVM linear', pipe1),
              ('KNN', pipe2),
              ('SVM RBF', pipe3)]
```

```python
stacking = StackingClassifier(estimators=estimators, final_estimator=final_estimator)
stacking.fit(X_train, y_train)
pred = stacking.predict(X_test)
print(classification_report(y_test, pred, digits=4))
```
