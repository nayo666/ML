# PCA

```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.preprocessing import StandardScaler
from sklearn import decomposition
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from random import randint
```

# Leer los datos
Leer el archivo covid.csv y mostrarlo

```python
covid = pd.read_csv("/home/noxd/Doc/Machine Learning/datos/covid.csv", header=0)
covid.head()
```

Poner los datos menos la columna Country en X y Country en la variable y (como arrays de numpy)

```python
X = covid.drop(["Country"], axis=1).to_numpy()
y = covid["Country"].to_numpy()
```

Antes de usar PCA, debemos tener los datos centrados en cero.

Usar StandardScaler para normalizar los datos.

Usar np.mean() para comprobar que los datos tengan promedio cero.

```python
scaler = StandardScaler()
(scaler.fit(X))
(scaler.transform(X))
(np.mean(scaler.transform(X)))
```

```python
markers = ['s', 'x', '+', '^']
```

Usar PCA con una componente en los datos.

Guardar los datos transformados en la variable X_t

```python
X_t.shape
```

```python
# Su código aquí
xt=(scaler.transform(X))

pca = PCA(n_components=1)
pca.fit(xt)
X_t = pca.transform(xt)

# Fin de código
plt.figure(figsize=(10, 2))
for i in range(21):
    plt.scatter(X_t[i], [0], marker=markers[i%4])
plt.legend(y, bbox_to_anchor=(1.05, 1), loc='upper left', ncol=3)
plt.show()
```

En el gráfico anterior podemos ver que USA, India y Brasil están separados del resto

Hacer los mismo para PCA con dos componentes, graficar los puntos como se hizo anteriormente.

```python
pca = PCA(n_components=2)
pca.fit(xt)
X_t = pca.transform(xt)

# Fin de código
plt.figure(figsize=(10, 6))
for i in range(21):
    plt.scatter(X_t[i, 0], X_t[i, 1], marker=markers[i%4])
plt.legend(y,  bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

Hacer los mismo para PCA con tres componentes y graficar.

```python
pca = PCA(n_components=3)
pca.fit(xt)
X_t = pca.transform(xt)
```

```python
fig = plt.figure(figsize=(15, 10))
ax = fig.add_subplot(111, projection='3d')


# plot 
colors = []
colors = ['r', 'b', 'y', 'c',"g","m","dimgray","darkorange","teal","deeppink","chocolate","tomato","palegreen","gold","indigo","lime","royalblue","darkgreen","tan","yellowgreen","orchid"]

for i in range(21):
    ax.scatter(X_t[i, 0], X_t[i, 1],  X_t[i, 2], c=[colors[i]], label=str(y[i]), marker=markers[i%4])

ax.legend(  ncol=2)
plt.show()
```
